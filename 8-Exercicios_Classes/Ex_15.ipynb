```python
class Bichinho:
    # Inicialização dos atributos da classe
    def __init__(self, nome, fome = 5, saude = 5, idade = 1):
        self.set_nome(nome)
        self.fome = fome
        self.saude = saude
        self.idade = idade
    
    def set_nome(self, nome):
        self.nome = nome
    
    # Método para alimentar o bichinho (quanto mais comida ele come menos fome ele tem)
    def alimenta(self, comida):
        if (self.fome - comida) < 0:
            self.fome = 0
        else:
            self.fome = self.fome - comida
    
    # Método para brincar com o bichinho (quanto mais ele brinca, mais saudável fica)
    def brinca(self, tempo):
        self.saude += tempo
        
    # Métodos para retornar os valores dos atributos
    def get_nome(self):
        return self.nome
    
    def get_fome(self):
        return self.fome
    
    def get_saude(self):
        return self.saude
    
    def get_idade(self):
        return self.idade
    
    # Método para calcular o humor do bichinho
    def humor(self):
        if self.fome == 0:
            self.fome = 1
        
        # É calculada uma porcentagem do nível da fome e do nível de saúde
        percent_saude = (100 * self.saude) / 10
        percent_fome = (100 * self.fome) / 10
        
        # Soma-se as duas porcentagens com a ressavla que a fome deve estar negativa pois quanto menos fome melhor
        percent_total = percent_saude + (percent_fome * -1)
        
        # Validação do nível do humor 
        if percent_total < 0:
            humor = 'Triste'
        elif percent_total == 0:
            humor = 'Neutro'
        else:
            humor = 'Feliz'
        
        return humor
```

```python
# Função que apresenta o menu
def menu():
    while True:
        print('''*----- Bichinho Virtual -----*
    1 - Criar Bichinho
    2 - Status do Bichinho
    3 - Ações com o Bichinho
    4 - Sair\n''')
        
        opcao = int(input('Digite uma opção: '))
        
        # Escolha da opção do menu
        if opcao == 1:
            # Foi utilizado uma variável global para poder acessá-la de fora da função
            global bichinho
            # Chamada da função que cria o objeto
            bichinho = cria_bichinho()
        
        # Status do bichinho
        elif opcao == 2:
            # Verificação se objeto existe
            try:
                status()
            
            # Caso não exista é apresntado a mensagem e redirecionado para a criação do objeto
            except NameError:
                print('Não existe um bichinho criado. Portanto será criado um agora!\n')
                bichinho = cria_bichinho()
                status()
        
        # Chamada da função do novo menu de ações do bichinho
        elif opcao == 3:
            menu_acoes()
            
        # Sair do programa
        elif opcao == 4:
            break
        
        else:
            print('Opção inválida!')
```

```python
# Criação do objeto
def cria_bichinho():
    nome = input('Digite o nome do bichinho: ')
    return Bichinho(nome)
```

```python
def status():
    print(f'''
    Nome: {bichinho.get_nome()}
    Fome: {bichinho.get_fome()}
    Saúde: {bichinho.get_saude()}
    Idade: {bichinho.get_idade()}
    Humor: {bichinho.humor()}\n''')
```

```python
# Menu de ações do bichinho
def menu_acoes():
    while True:
        print('''*----- Ações -----*
    1 - Alimentar
    2 - Brincar
    3 - Voltar ao menu inicial''')
        opcao = int(input('Digite uma opção: '))
        
        if opcao == 1:
            comida = int(input('Digite a quantidade de comida a ser dada: '))
            # limitação do maior valor a ser inserido para atuar na porcentagem do humor
            if comida > 10:
                comida = 10
            
            # O método alimenta recebe sempre o valor positivo
            bichinho.alimenta(abs(comida))
            
        elif opcao == 2:
            tempo = int(input('Digite a quantidade de minutos a se brincar: '))
            # Validação para limitar o tempo a minutos
            if tempo > 59:
                tempo = 59
            
            # A cada 10 minutos aumenta-se um ponto na saúde
            bichinho.brinca(abs(round(tempo/10)))
            
        elif opcao == 3:
            break
        else:
            print('Opção inválida!')
```
